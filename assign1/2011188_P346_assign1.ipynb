```python
# sum of 1st N odd numbers

i=s=1
j=0
x=int(input("enter no. of terms: "))
while (i<x):
  j = 2*i + 1
  s = s + j
  i = i + 1
print("sum =", s)
```

```python
# factorial of N

i=f=1
x=int(input("enter no.: "))
while (i<=x):
  f = f*i
  i = i + 1
print("factorial =", f)
```

```python
# sum of N terms of an AP, GP and HP

# AP
a = int(input("enter 1st no.: "))
n = int(input("enter the no. of terms: "))
d = 1.5

total = 0
value = a
for i in range(n):
    total = total + value
    value = value + d
    
print("\nthe sum of AP series =", total)
```

```python
# GP
a = int(input("enter 1st no.: "))
n = int(input("enter the no. of terms: "))
r = 0.5

total = 0
value = a
for i in range(n):
    total = total + value
    value = value * r
    
print("\nthe sum of GP series =", total)
```

```python
# HP #
a = int(input("enter 1st no.: "))
n = int(input("enter the no. of terms: "))
d = 1.5
total = 0
value = a

for i in range(n-1):
  reci = 1/value
  total += reci
  value += d
    

print("\nthe sum of HP series =", total)
```

```python
# graph of sum of series vs. n
import matplotlib.pyplot as plt
x=[]
y=[]
n=20
total = 0
value = 0
for i in range(n):
  value = (-1)**(i+1)/2**i
  total = total + value
  x.append(i)
  y.append(total)

plt.plot(x,y)
print("n =", x)
print("sum =", y)
```

```python
class complex ():
    def initcomplex(self):
        self.realPart = int(input("enter the real part: "))
        self.imgPart = int(input("enter the imaginary part: "))            

    def display(self):
        print(self.realPart,"+",self.imgPart,"i", sep="")

    def sum(self, c1, c2):
        self.realPart = c1.realPart + c2.realPart
        self.imgPart = c1.imgPart + c2.imgPart

    def product(self, c1, c2):
      self.realPart = c1.realPart*c2.realPart + (-1)*c1.imgPart*c2.imgPart
      self.imgPart = c1.realPart*c2.imgPart + c2.realPart*c1.imgPart
    
    def modulus(self):
      print(((self.realPart**2) + (self.imgPart**2))**(1/2))

c1 = complex()
c2 = complex()
c3 = complex()
c4 = complex()

print("enter first complex number")
c1.initcomplex()
print("first complex number: ", end="")
c1.display()

print("enter second complex number")
c2.initcomplex()
print("second complex number: ", end="")
c2.display()

print("sum of 2 complex numbers is: ", end="")
c3.sum(c1,c2)
c3.display()

print("product of 2 complex numbers is: ", end="")
c4.product(c1,c2)
c4.display()

print("modulus of 2 complex numbers are: ")
c1.modulus()
c2.modulus()
```
