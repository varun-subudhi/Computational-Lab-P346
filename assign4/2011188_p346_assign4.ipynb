```python
def fn2(x):
  return -x-cos(x)

print("Roots in the following methods:\n")
r1, e1, i1  = bisection(fn2, -1, 1, 1e-6)
print("Bisection: ", r1)

r2, e2, i2 = regula_falsi(fn2, -1, 1, 1e-6)
print("Regula-falsi:", r2)

r3, e3, i3 = newton_raphson(fn2, -1, 1e-6)
print("Newton-Raphson:", r3)
```
