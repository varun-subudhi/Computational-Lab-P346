```python
from library import *
import math

A = 3.14159*2*1 
j = 0
x = 0.1
n = 10**5
for i in range(n):
    r1 = myrand(x)
    x = r1/2
    r2 = myrand(x)
    x = r2
    if (r1**2) + (r2**2) <= 1:
        j += 1
a = 8*j/n
print("Area of ellipse using LCG: ", a)
print("Analytical value of area of ellipse: ", A)

error = (abs(A-a)/A)*100
print("Error in calculating area: ", error, "%")
```
