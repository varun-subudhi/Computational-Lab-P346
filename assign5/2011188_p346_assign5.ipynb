We know, for a beam whose mass density depends on the length of beam, center of mass can be calculated as:

$x_{cm} = \int{xdm}/\int{dm}$ ; where dm = $x^2$dx

We will use Simpson Integration Scheme as it provides the most accurate accurate approximation of the integral.

```python
# 3

def xdm(x):
  return x**3

def dm(x):
  return x**2

xcm = simpson(xdm,0,2,1000000)/simpson(dm,0,2,1000000)

print("The center of the mass of the given beam is at", xcm, "cm from its (x=0) end.")
```
